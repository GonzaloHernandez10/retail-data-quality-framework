# AUDITORÍA DE INTEGRIDAD ESTRUCTURAL (Profiling)

  - Verificar que las columnas críticas (quantity, unit_price, order_date, category, order_id) existan y tengan el tipo de dato correcto (Numérico, Fecha, String).

  - Identificar registros con nulos en cualquier columna del dataset.


```python
####################################
# Verificación de columnas criticas
####################################
def verificar_columnas_criticas(df, columnas_criticas):
  for columna in columnas_criticas:
    tipo_columna = df[columna].dtype
    if columna in df.columns:
      if columna == 'order_date' and tipo_columna == 'object':
        print(f'⚠️ AVISO: {columna} es "{tipo_columna}", se recomienda convertir a datetime.')
      else:
        print(f'✔️ OK: {columna} es "{tipo_columna}"')

columnas_criticas = ['quantity', 'unit_price', 'order_date', 'category', 'order_id']
verificar_columnas_criticas(df, columnas_criticas)
```

```python
##################################################
# Cuantificación del volumen inicial de registros
##################################################
total_registros_iniciales = len(df.index)
print('✅ El total inicial de registros es:', total_registros_iniciales)
```

```python
##################################
# Identificación de valores nulos
##################################
def identificar_valores_nulos(df):
  valores_nulos = df.isnull().any()
  return valores_nulos

valores_nulos = identificar_valores_nulos(df)
valores_nulos
```

```python
#################################################
# Identificación del porcentaje de valores nulos
#################################################
valores_nulos = df.isna().sum()
porcentaje_valores_nulos = (valores_nulos / len(df)) * 100

resumen_nulos = pd.DataFrame({
  'Valores nulos': valores_nulos,
  'Porcentaje de valores nulos': porcentaje_valores_nulos
})
resumen_nulos

print(f'Resumen nulos: {resumen_nulos[resumen_nulos['Valores nulos'] > 0]}')
```

# EJECUCICIÓN DE REGLAS DE VALIDACIÓN (Validation)

  - Validación de integridad: Identificar registros con valores nulos en los cinco campos críticos (quantity, unit_price, order_date, category, order_id).

  - Validación de negativos: Detectar registros donde quantity o unit_price sean iguales o menores a 0.

  - Validación de rango: Identificar registros donde el discount sea mayor a 0.35 o menor a 0.

  - Validación de unicidad: Detectar registros que compartan los mismos atributos clave (duplicados exactos).

  - Validación lógica: Detectar registros con valores nulos o faltantes en los campos discount y region.

```python
###########################
# Validación de integridad
###########################
columnas_criticas = ['quantity', 'unit_price', 'order_date', 'category', 'order_id']
validacion_integridad = df[df[columnas_criticas].isna().any(axis=1)]
validacion_integridad.info()
```

```python
###########################
# Validación de negativos
###########################
maks = (df['quantity'] <= 0) | (df['unit_price'] <= 0)
validacion_negativos = df[maks]
validacion_negativos.info()
```

```python
###########################
# Validación de rango
###########################
mask = (df['discount'] > 0.35) | (df['discount'] < 0)
validacion_rango = df[mask]
validacion_rango.info()
```

```python
#########################
# Validación de unicidad
#########################
mask = df.duplicated(subset=['order_id', 'quantity', 'unit_price', 'order_date', 'category'])
validacion_unicidad = df[mask]
validacion_unicidad.info()
```

```python
#########################
# Validación de lógica
#########################
mask = (df['discount'].isna()) | (df['region'].isna())
validacion_logica = df[mask]
validacion_logica.info()
```

# ClASIFICACIÓN Y ETIQUETADO (Tagging)

  - Asignar un código de error específico a cada registro que falle en alguna de las validaciones anteriores.

  - Categorizar cada registro en uno de los tres estados:

    - VALID: Supera todas las validaciones.

    - CORRIGIBLE: Falla solo en validaciones de descuento o campos no críticos.

    - NON-RELIABLE: Falla en 1 o más campos críticos o tiene múltiples errores.

**Sistema de "Error Tags"** (Taxonomía de Errores):

| CÓDIGO DE ERROR | DESCRIPCIÓN | TIPO | ACCIÓN
|:---:|:---|:---:|:---|
| CRITICAL_ERROR | Fallo en campo crítico o valores negativos en cantidad o precio. | Crítico | Eliminar del análisis |
| INVALID_RANGE | Descuento fuera de rango (>35%). | No crítico | Corregir (Imputar 0 o max) |
| DUPLICATED | Transacción idéntica duplicada | Duplicado | Conservar solo uno |
| INVALID_LOGIC | Inconsistencia | No crítico | Corregir (Inputar a 0% o 'Unknown')

<br>

```python
################################################################
# Creación de las columnas category_value, error_code, is_valid
################################################################
df['category_value'] = 'VALID'
df['error_code'] = 'NONE'
df['is_valid'] = True
```

```python
############################################################
# Identificación de errores corregibles y de BAJA PRIORIDAD
############################################################
mask_corregible = (df['region'].isna() | df['discount'].isna())
df.loc[mask_corregible, 'category_value'] = 'CORRIGIBLE'
df.loc[mask_corregible, 'error_code'] = 'INVALID_LOGIC'
```

```python
############################################################
# Identificación de errores críticos y de ALTA PRIORIDAD
############################################################
mask_critico = (
  df[['quantity', 'unit_price', 'order_date', 'category', 'order_id']].isna().any(axis=1) |
  (df['quantity'] <= 0) |
  (df['unit_price'] <= 0)
)

df.loc[mask_critico, 'category_value'] = 'NON-RELIABLE'
df.loc[mask_critico, 'error_code'] = 'CRITICAL_ERROR'
df.loc[mask_critico, 'is_valid'] = False
```

```python
#############################################
# Tramiento especial para valores duplicados
#############################################
df.loc[df.duplicated(keep='first'), 'error_code'] = 'DUPLICATED'
```

```python
print(df['category_value'].value_counts())
print('\n')
print(df['error_code'].value_counts())
print('\n')
print(df['is_valid'].value_counts())
```

# GESTIÓN DE REGISTROS (Actioning)

  - Preservación: Separar los registros "Válidos" para el dataset final.

  - Imputación: Aplicar la lógica de corrección a los "Corregibles" (ej. convertir discount nulo a 0).

  - Exclusión: Aislar los registros "No Confiables" en un repositorio de errores para auditoría.

  - Deduplicación: Eliminar las instancias repetidas de los registros duplicados, conservando solo una.

```python
##################################
# Separación de registros válidos
##################################
valores_validos = df[df['category_value'] == 'VALID']
valores_validos.sample(50)
```

```python
valores_validos.info()
```

```python
valores_validos.isna().any()
```

```python
###########################################################
# Aplicación de reglas de imputación a valores corregibles
###########################################################
valores_corregibles = df[df['category_value'] == 'CORRIGIBLE'].copy()
valores_corregibles.head(50)
```

```python
valores_corregibles['discount'] = np.where(valores_corregibles['discount'].isna(), 0, valores_corregibles['discount'])
valores_corregibles.head(50)
```

```python
valores_corregibles['region'] = np.where(valores_corregibles['region'].isna(), 'Unknown', valores_corregibles['region'])
valores_corregibles.head(50)
```

```python
valores_corregibles.info()
```

```python
valores_corregibles.isna().any()
```

```python
#############################################
# Aislamiento de los registros no confiables
#############################################
valores_no_confiables = df[df['category_value'] == 'NON-RELIABLE']
valores_no_confiables.head(50)
```

```python
########################################################################
# Concatenación de los dataframes valores_validos y valores_corregibles
########################################################################
df_final = pd.concat([valores_validos, valores_corregibles]).reset_index(drop=True)
df_final.head(50)
```

```python
df_final.info()
```

```python
df_final['category_value'].value_counts()
```

```python
########################################
# Eliminación de los valores duplicados
########################################
df_final.drop_duplicates(subset=['quantity', 'unit_price', 'order_date', 'category', 'order_id'], keep='first', inplace=True)
df_final.info()
```

# REINGENIERÍA DE CÁLCULO (Recalculation)

  - Calcular gross_sales (quantity * unit_price) para todos los registros válidos y corregidos.

  - Calcular net_sales aplicando el descuento validado sobre las gross_sales.

  - Comparar los nuevos cálculos contra los valores originales para detectar desviaciones técnicas.

```python
############################
# Re-cálculo de gross_sales
############################
df_final['gross_sales_fixed'] = df_final['quantity'] * df_final['unit_price']
df_final.head(50)
```

```python
##########################
# Re-cálculo de net_sales
##########################
df_final['net_sales_fixed'] = df_final['gross_sales_fixed'] * (1 - df_final['discount'])
df_final.head()
```

```python
##########################
# Análisis de impacto
##########################
if 'gross_sales' in df_final.columns:
  df_final['diff_gross'] = df_final['gross_sales_fixed'] - df_final['gross_sales']
  print(f"Desviación técnica detectada en gross Sales: {df_final['diff_gross'].sum()}")
```

# GENERACIÓN DE EVIDENCIAS (Reporting)

  - Obtener ventas brutas iniciales
  - Obtener ventas brutas válidada
  - Obtener el dinero en riesgo

```python
#############################################
# Obtención de las ventas brutas iniciales
#############################################
ventas_brutas_iniciales = df['gross_sales'].sum()
ventas_brutas_iniciales
```

```python
#########################################
# Obtención de las ventas brutas válidas
#########################################
ventas_brutas_validad = df_final['gross_sales_fixed'].sum()
ventas_brutas_validad
```

```python
#######################
# Obtención del riesgo
#######################
riesgo = ventas_brutas_iniciales - ventas_brutas_validad
riesgo

# Cálculo del porcentaje de desviación
porcentaje_desviacion = (riesgo / ventas_brutas_iniciales) * 100

print(f"Riesgo mitigado: ${riesgo:,.2f} ({porcentaje_desviacion:.2f}% de desviación corregida sobre el reporte original)")
```

```python
###############################
# Generación del Bridge Report
###############################
bridge_report = pd.DataFrame({
  'Ventas brutas iniciales': [ventas_brutas_iniciales],
  'Ventas brutas válidas': [ventas_brutas_validad],
  'Dinero en riesgo': [riesgo]
})

bridge_report
```

# ANÁLISIS DE VENTAS CONFIABLE

  - Obtener ventas por categoría: ¿Cuál es la categoría que más dinero genera realmente?

  - Obtener ventas por región: ¿Hay regiones que parecen rendir menos ahora que eliminamos los datos basura?

  - Obtener ticket promedio (AOV):

```python
import matplotlib.pyplot as plt
import seaborn as sb
from matplotlib.ticker import StrMethodFormatter
```

```python
plt.style.use('seaborn-v0_8-whitegrid')
```

```python
#############################################
# Obtención de gráfica ventas por categorías
#############################################
df_ventas_por_categoria = df_final.groupby('category')['gross_sales_fixed'].sum().reset_index()
df_sorted = df_ventas_por_categoria.sort_values(by='gross_sales_fixed', ascending=False);

plt.figure(figsize=(10,6));
sb.barplot(data=df_sorted, x='category', y='gross_sales_fixed', palette='viridis');
plt.title('Ventas Netas por Categoría (Datos Validados)');
plt.xticks(rotation=45); # Por si los nombres de categoría son largos
plt.gca().yaxis.set_major_formatter(StrMethodFormatter('${x:,.0f}'))
plt.show();
```

```python
#########################################
# Obtención de gráfica ventas por región
#########################################
df_ventas_por_region = df_final.groupby('region')['gross_sales_fixed'].sum().reset_index()
df_sorted = df_ventas_por_region.sort_values(by='gross_sales_fixed', ascending=False);

plt.figure(figsize=(10,6));
sb.barplot(data=df_sorted, x='region', y='gross_sales_fixed', palette='viridis');
plt.title('Ventas Netas por Región (Datos Validados)');
plt.xticks(rotation=45); # Por si los nombres de categoría son largos
plt.gca().yaxis.set_major_formatter(StrMethodFormatter('${x:,.0f}'))
plt.show();
```

```python
####################
# Obtención del aov
####################
aov = df_final['net_sales_fixed'].mean()
print(f"El Ticket Promedio (AOV) validado es: ${aov:,.2f}")
```

```python
########################################################
# Obtención de los datos para el análisis de causa raíz
########################################################

# Contabilizamos cuántos errores hubo de cada tipo
rca_counts = valores_no_confiables['error_code'].value_counts()
rca_percent = (valores_no_confiables['error_code'].value_counts(normalize=True) * 100)

# Verificamos en qué categorías se concentran los errores
rca_by_category = valores_no_confiables.groupby('category')['error_code'].count().sort_values(ascending=False)

print("--- Distribución de Errores (Causa Raíz) ---")
print(rca_counts)
print("\n--- Distribución de Errores en % (Causa Raíz) ---")
print(rca_percent)
print("\n--- Categorías con mayor incidencia de errores ---")
print(rca_by_category)
```
